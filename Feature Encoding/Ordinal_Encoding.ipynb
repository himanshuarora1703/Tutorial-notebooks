```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/customer.csv')
df
```

```python
# check the info of the dataframe

df.info()
```

```python
# make x and y

X = df.drop(columns='purchased')
y = df['purchased']
```

```python
# Do train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

X_train.shape,X_test.shape
```

```python
# check the value counts in columns

for col in df.columns[1:4]:
  print(col,df[col].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder,OneHotEncoder
```

```python
# do the transformations

from sklearn.compose import ColumnTransformer

ord = OrdinalEncoder(categories=[['Poor','Average','Good'],['School', 'UG', 'PG']],dtype=np.int16)
ohe= OneHotEncoder(drop='first',sparse=False,dtype=np.int16)
trans = ColumnTransformer(transformers=[
    ('ordinal',ord,['review','education']),
    ('ohe',ohe,['gender'])
])
```

```python
X_train_trans = trans.fit_transform(X_train)
X_test_trans = trans.transform(X_test)
```

```python
# encode the target

le = LabelEncoder()
y_train_trans = le.fit_transform(y_train)
y_test_trans = le.transform(y_test)

y_test_trans

le.classes_
```
