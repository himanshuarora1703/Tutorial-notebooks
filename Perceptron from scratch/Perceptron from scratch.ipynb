```python
# import the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
# Generate a dataset
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
# plot a scatter plot

plt.scatter(X[:,0],X[:,1],c=y,marker='x')
```

```python
def my_perceptron(X,y):
    lr = 0.01
    epoch = 1000
    
    X = np.insert(X,0,1.0,axis=1)
    weights = np.ones(X.shape[1])
    
    
    for i in range(epoch):
        random = np.random.randint(low=0,high=100)
        y_hat = np.dot(X[random],weights)
        
        if y_hat>0:
            y_hat = 1
        else:
            y_hat = 0
        
        weights = weights + (lr*(y[random]-y_hat))*X[random]
        
    return weights[0],weights[1:]
```

```python
intercept,coef = my_perceptron(X,y)
```

```python
m = -(coef[0]/coef[1])
c = -(intercept/coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
# plot the graph

plt.scatter(X[:,0],X[:,1],c=y,marker='x')
plt.plot(x_input,y_input,color='red')
plt.ylim(-4,3)
```
