```python
# problem in vectorized opertions in vanilla python
s = ['cat','mat',None,'rat']

[i.startswith('c') for i in s]
```

```python
# How pandas solves this issue?

s = pd.Series(['cat','mat',None,'rat'])
# string accessor
s.str.startswith('c')
s[s.str.startswith('c').fillna(False)]
# fast and optimized
```

```python
# import titanic
df = pd.read_csv('/content/titanic (1).csv')
df['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
# strip
"                   nitish                              ".strip()
df['Name'].str.strip()
```

```python
# passenger with the largest name

df.loc[df['Name'].str.len() == df['Name'].str.len().max(),'Name']
```

```python
# split -> get
df['lastname'] = df['Name'].apply(lambda x : x.split(',')[0])
df['firstname'] = df['Name'].str.split(',').str.get(1)
df.head()
```

```python
df[['Title','firstname']] = df['Name'].str.split(',').str.get(1).str.lstrip().str.split(' ',n=1,expand=True)

df
```

```python
df['Title'].value_counts()
```

```python
# replace
df['Title'] = df['Title'].str.replace('Ms.','Miss.')
df['Title'] = df['Title'].str.replace('Mlle.','Miss.')
```

```python
df['Title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.startswith('A')]
```

```python
# applying regex
# contains
# search john -> both case
df[df['firstname'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```

```python
# slicing
df['Name'].str[::2]
```
